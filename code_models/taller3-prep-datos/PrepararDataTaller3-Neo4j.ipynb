```python
import pandas as pd
```

```python
df_ratings = pd.read_csv("ratings.csv")
```

```python
df_ratings.head(10)
```

```python
df_user_mean=df_ratings.groupby('userId')['rating'].mean().reset_index()
df_user_mean.columns=['userId','mean']
df_user_mean
```

```python
df_ratings= pd.merge(df_ratings,df_user_mean,how="inner",left_on="userId", right_on="userId")
```

```python
df_ratings.head(10)
```

```python
df_ratings_likes = df_ratings[(df_ratings["rating"] >= df_ratings["mean"]) & (df_ratings["rating"] >= 3)]
```

```python
df_ratings_likes.head(10)
```

```python
df_tag_relevance = pd.read_csv("genome-scores.csv")
```

```python
df_tag_relevance.head(10)
```

```python
df_tag_ids = pd.read_csv("genome-tags.csv")
```

```python
df_tag_ids.head(10)
```

```python
df_tags_info= pd.merge(df_tag_relevance,df_tag_ids,how="inner",left_on="tagId", right_on="tagId")
```

```python
df_tags_info.head(10)
```

```python
df_tags_info[df_tags_info.movieId==1].sort_values(by=['relevance'], ascending=False).head(10)
```

```python
df_tag_user = pd.read_csv("tags.csv")
```

```python
df_tag_user.head(10)
```

```python
df_tag_user[df_tag_user.movieId==1].tag.value_counts().head(10)
```

```python
df_tag_user_count = df_tag_user.groupby(['movieId', 'tag']).size().reset_index(name='counts')
```

```python
df_tag_user_count[df_tag_user_count.movieId==1].sort_values(by=['counts'], ascending=False).head(10)
```

```python
df_tags_total = pd.merge(df_tag_user_count,df_tags_info,how="inner",left_on=["movieId","tag"], right_on=["movieId","tag"])
```

```python
df_tags_total.head(10)
```

```python
df_tags_total.isna().sum()
```

```python
df_tags_total['score'] = df_tags_total['counts'] + df_tags_total['relevance']
```

```python
df_tags_total.head(10)
```

```python
df_tags_total[df_tags_total.movieId==1].sort_values(by=['score'], ascending=False).head(10)
```

```python
df_tag_user_count[df_tag_user_count.movieId==1].sort_values(by=['counts'], ascending=False).head(10)
```

```python
df_tags_total_firts5 = df_tags_total.sort_values(['movieId','score'],ascending=False).groupby('movieId').head(5)
df_tags_total_firts5.head(100)
```

```python
df_tags_total_firts5_row = df_tags_total_firts5.groupby('movieId')['tag'].apply(list).reset_index(name='tags')
df_tags_total_firts5_row.head(10)
```

```python
len(df_movies)
```

```python
len(df_tags_total_firts5_row)
```

```python
df_movies_neo4j = df_movies.merge(df_tags_total_firts5_row, how='left', on='movieId')
df_movies_neo4j
```

```python
mask = df_movies_neo4j['tags'].notnull()
df_movies_neo4j.loc[mask, 'tags'] = ['|'.join(map(str, x)) for x in df_movies_neo4j.loc[mask, 'tags']]
df_movies_neo4j
```

```python
df_ratings_likes_neo4j = df_ratings_likes[["userId","movieId","rating"]]
df_ratings_likes_neo4j
```

```python
df_movies_diract_orig = pd.read_csv("diract.csv")
df_movies_diract_orig = df_movies_diract_orig.sort_values(by=['movieId'], ascending=True)
df_movies_diract_orig.reset_index(inplace=True)
df_movies_diract_orig.drop('index', axis = 1, inplace=True)
```

```python
df_movies_diract_orig
```

```python
df_movies_neo4j
```

```python
df_movies_neo4j["directores"] = df_movies_diract_orig["directores"]
df_movies_neo4j["actores"] = df_movies_diract_orig["actores"]
```

```python
df_movies_neo4j
```

```python
df_movies_neo4j_csv = df_movies_neo4j.rename(columns={"movieId": "movieId:ID"})
df_movies_neo4j_csv["movieId:ID"] = "m"+df_movies_neo4j_csv["movieId:ID"].astype(str)
df_movies_neo4j_csv
```

```python
df_movies_neo4j_csv.to_csv("movies_neo4j.csv",index=False,sep = ",")
```

```python
df_user_mean_csv = df_user_mean.rename(columns={"userId": "userId:ID"})[["userId:ID"]]
df_user_mean_csv["userId:ID"] = "u"+df_user_mean_csv["userId:ID"].astype(str)
df_user_mean_csv.head(10)
```

```python
df_user_mean_csv.to_csv("users_neo4j.csv",index=False,sep = ",")
```

```python
df_ratings_likes_neo4j
```

```python
df_ratings_likes_neo4j_csv = df_ratings_likes_neo4j.rename(columns={"userId": ":START_ID","movieId" : ":END_ID", "rating" : "rating:double"})
df_ratings_likes_neo4j_csv[":START_ID"] = "u"+df_ratings_likes_neo4j_csv[":START_ID"].astype(str)
df_ratings_likes_neo4j_csv[":END_ID"] = "m"+df_ratings_likes_neo4j_csv[":END_ID"].astype(str)
df_ratings_likes_neo4j_csv = df_ratings_likes_neo4j_csv[[":START_ID","rating:double",":END_ID"]]
df_ratings_likes_neo4j_csv.head(10)
```

```python
df_ratings_likes_neo4j_csv.to_csv("ratings_neo4j.csv",index=False,sep = ",")
```
